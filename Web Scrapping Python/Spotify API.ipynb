```python
#Authentication - without user
client_credentials_manager = SpotifyClientCredentials(client_id=cid, client_secret=secret)
sp = spotipy.Spotify(client_credentials_manager = client_credentials_manager)
```

# Get Data

```python
playlist_link = "https://open.spotify.com/playlist/2S2bAjLaE0CN2V0cTwAeEL?si=fc51cd910be5436f"
playlist_URI = playlist_link.split("/")[-1].split("?")[0]
track_uris = [x["track"]["uri"] for x in sp.playlist_tracks(playlist_URI)["items"]]
```

```python
playlist_track_uri = []
playlist_track_name = []
playlist_artist_uri = []
playlist_artist_info = []
playlist_artist_popularity = []
playlist_artist_genre = []
playlist_album = []
playlist_track_popularity = []
playlist_artist_name= []
playlist_track_id= []
```

```python
for track in sp.playlist_tracks(playlist_URI)["items"]:
    # ID
    track_id = track['track']['id']
    playlist_track_id.append(track_id)

    #URI
    track_uri = track["track"]["uri"]
    playlist_track_uri.append(track_uri)

    #Track name
    track_name = track["track"]["name"]
    playlist_track_name.append(track_name)

    #Main Artist
    artist_uri = track["track"]["artists"][0]["uri"]
    playlist_artist_uri.append(artist_uri)
    artist_info = sp.artist(artist_uri)
    playlist_artist_info.append(artist_info)
    
    #Name, popularity, genre
    artist_name = track["track"]["artists"][0]["name"]
    playlist_artist_name.append(artist_name)
    artist_pop = artist_info["popularity"]
    playlist_artist_popularity.append(artist_info)
    artist_genres = artist_info["genres"]
    playlist_artist_genre.append(artist_genres)
    
    #Album
    album = track["track"]["album"]["name"]
    playlist_album.append(album)

    #Popularity of the track
    track_pop = track["track"]["popularity"]
    playlist_track_popularity.append(track_pop)
```

# Extract

```python
data_mentah = sp.audio_features(playlist_track_id)
```

```python
data = pd.DataFrame(data=data_mentah,columns=data_mentah[0].keys())
```

```python
data.head()
```

```python
data['track_name']=playlist_track_name 
data['artist_uri']=playlist_artist_uri 
data['genre']=playlist_artist_genre
data['track_popularity']=playlist_track_popularity
data['artist_name']=playlist_artist_name
```

```python
data.info()
```

# Cleaning

```python
data['uri']=data['uri'].str.replace('spotify:track:','')
data['artist_uri']=data['artist_uri'].str.replace('spotify:artist:','')
data['genre']=data['genre'].astype('str')
data['genre']=data['genre'].str.replace("]", "")
data['genre']=data['genre'].str.replace("[", "")
```

```python
data=data.drop(columns='type',axis=1)
```
